## Create the config YAML file

```python
import os
PROJ_DIR = "/content/drive/Othercomputers/laptop/jupyter"
CONFIG_DIR = os.path.join(PROJ_DIR, "configs")
CONFIG_FILE = os.path.join(CONFIG_DIR, "default.yaml")
if not os.path.exists(CONFIG_DIR):
    os.mkdir(CONFIG_DIR)

config_text = \
"""image_size: 600x600
num_classes: 7
label_map: {1: ottavaBracket, 2: dynamicCrescendoHairpin, 3: dynamicDiminuendoHairpin, 4: slur,  5: tie, 6: tupletBracket,6: brace,7: beam}
input_rand_hflip: false
jitter_min: 1
jitter_max: 1
max_instances_per_image: 50
"""

with open(CONFIG_FILE, "w") as fwrite:
    fwrite.write(config_text)

MODEL = "efficientdet-d1"
CKPT = MODEL


TRAIN_SET = os.path.join(PROJ_DIR, "data/whole/train", "record*")
VAL_SET = os.path.join(PROJ_DIR, "data/whole/val", "record*")
MODEL_DIR_TMP = os.path.join(PROJ_DIR, "tmp", f"{MODEL}-finetune")
TRAIN_NUM_EXAMPLES = 1000*len(os.listdir("data/whole/train"))
EVAL_NUM_EXAMPLES = 1000*len(os.listdir("data/whole/val"))
EPOCHS = 1
BATCH_SIZE = 8
```

```python
cd /content/drive/Othercomputers/laptop/jupyter/automl/efficientdet
```

```python
if os.path.exists(MODEL_DIR_TMP):
    !rm -rf {MODEL_DIR_TMP}
!pip install PyYaml==5.1
!python -m main \
    --mode=train_and_eval \
    --train_file_pattern={TRAIN_SET} \
    --val_file_pattern={VAL_SET} \
    --model_name={MODEL} \
    --ckpt={CKPT} \
    --model_dir={MODEL_DIR_TMP} \
    --train_batch_size={BATCH_SIZE} \
    --eval_batch_size={BATCH_SIZE} \
    --num_epochs={EPOCHS} \
    --num_examples_per_epoch={TRAIN_NUM_EXAMPLES} \
    --eval_samples={EVAL_NUM_EXAMPLES} \
    --hparams={CONFIG_FILE}
```

```python
import os
MODEL_DIR_EXPORT = os.path.join(PROJ_DIR, "models/whole", f"{MODEL}-finetune")

MIN_SCORE_THRESHOLD = 0.5

!python -m model_inspect \
    --runmode=saved_model \
    --model_name={MODEL} \
    --ckpt_path={MODEL_DIR_TMP} \
    --saved_model_dir={MODEL_DIR_EXPORT} \
    --min_score_thresh={0.5} \
    --hparams={CONFIG_FILE}
```
